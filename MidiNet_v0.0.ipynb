### get listfile

```python
def get_listfile(dataset_path):
    list_file = []
    for root, dirs, files in os.walk(dataset_path):
        for f in files:
            if "MIDI-Unprocessed" in f and f.endswith(".midi"):
                fp = join(root, f)
                list_file.append(fp)
    return list_file

dataset_path = "/home/gloria/Scrivania/Neural networks and deep learning/Project/MidiNet-by-pytorch-Debugged/data/2004"
list_file = get_listfile(dataset_path)
list_file
```

## Check if 4/4 beats

```python
def beats_(list_):
    list_of_four_beat = []
    for file_ in list_:
        try:
            chorus_file = mido.MidiFile(file_)
            for track in chorus_file.tracks:
                for msg in track:
                    if msg.type == 'time_signature':
                        if msg.numerator == 4 and msg.denominator == 4:
                            list_of_four_beat.append(file_)
        except Exception as e:
            print(f"Cannot open or read the file {file_}: {e}")
    return list_of_four_beat

# Example usage:
files_with_4_4_time_signature = beats_(list_)
print("Files with 4/4 time signature:", files_with_4_4_time_signature)
```
